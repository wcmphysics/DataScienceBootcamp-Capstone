# Select and load specific data from SQL server

Create connection to server

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd

load_dotenv("../.env")
DB_STRING = os.getenv('DB_STRING')
db = create_engine(DB_STRING)
```

Get unique hdd models

```python
query_string = """SELECT model, COUNT(model) FROM "2021"
                WHERE date = '2021-10-01'
                GROUP BY model;"""
models = pd.read_sql(query_string, db).sort_values("count", ascending=False)
models
```

```python
query_string = """SELECT * FROM "2021"
                WHERE model = 'ST12000NM0008' AND date = '2021-10-10'
                LIMIT 1000;"""
df = pd.read_sql(query_string, db)
df.head()
```
